## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seaborn comes with built-in data sets!
sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
df.info()
```

##**Styling**

```python
sns.set_style('whitegrid')    #white, dark, whitegrid, darkgrid, ticks
plt.figure(figsize=(8,4))     #size
sns.set_context('paper', font_scale=2)    #paper, talk, poster


sns.stripplot(x='cylinders', y='mpg', data=df , jitter=True , hue='origin' , palette='magma')  #dodge separates hues in x

plt.legend(loc=2 , fontsize=9)  # upper right:1, upper left: 2 , lower left: 3 , lower right: 4

plt.xlabel('Cylinders', fontsize=16);
plt.ylabel('Mpg', fontsize=16);

plt.title('Cars', fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=10)    #value size on x and y
sns.despine(left=True , bottom=True)

plt.show()
```

##**Distribution - Histogram Plots**

####Distribution

```python
sns.set_style('whitegrid')    #white, dark, whitegrid, darkgrid, ticks
plt.figure(figsize=(8,4))


sns.displot(df['mpg'])
plt.show()
```

```python
sns.set_style('dark')    #white, dark, whitegrid, darkgrid, ticks
plt.figure(figsize=(8,4))

sns.displot(df['mpg'],kde=True,bins=20, color="red");
```

```python
df.head()
```

####DensityPlot

```python
sns.kdeplot(df['mpg'], color="red");
```

```python
sns.kdeplot(x=df['mpg'],y=df['cylinders'], color="green" );
```

####Scatter Plot

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg', y='horsepower', data=df, hue='cylinders', size='weight')
plt.show()
```

####Jointplot

```python
sns.jointplot(x='weight',y='mpg',data=df)
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=df ,kind='reg')
plt.show()
```

####Pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

```python
df.head(5)
```

```python
df1 = df[['mpg','cylinders','weight']]
df1.head(5)
```

```python
sns.pairplot(df1) # pairplot look at the joint relationship between numerical variables in the dataframe.
plt.show()
```

```python
df1['cylinders'].value_counts()
```

```python
sns.pairplot(df1,hue='cylinders',palette='coolwarm')
plt.show()
# hue is used for categorical variables
# palette specifies the colors
```

##**Categorical plots**

####Boxplot

```python
sns.boxplot(x='cylinders', y='mpg', data=df , hue='cylinders', showfliers=True)    #showfliers-Outliers
plt.legend(loc=1 , fontsize=8)
plt.show()
```

####Catplot

```python
sns.catplot(x='cylinders', y='mpg', data=df , hue='origin', kind='box')
plt.legend(fontsize=5)
plt.show()
```

####BarPlot

```python
sns.barplot(x='cylinders', y='mpg', data=df , estimator=np.mean)       #estimator=mean, median , std, var , cov
plt.show()
# barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean.

# the black vertical lines are the error bars.
```

####Countplot

```python
sns.countplot(x='cylinders',data=df)
plt.show()
```

####Violinplot

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='cylinders', y='mpg', data=df , hue='origin')
plt.legend(loc=2)
plt.show()
```

####Stripplot

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='cylinders', y='mpg', data=df , jitter=True , hue='origin', dodge=True)   #dodge separates hues in x

plt.legend(loc=1)
plt.show()
```

####Swarmplot

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='cylinders', y='mpg', data=df )

sns.swarmplot(x='cylinders', y='mpg', data=df , color='yellow' , hue='origin')
#sns.swarmplot(x='cylinders', y='mpg', data=df , hue='origin')

plt.legend(loc=1)
plt.show()
```

####Pointplot

```python
plt.figure(figsize=(10,8))
sns.pointplot(x='cylinders', y='mpg', data=df ,  hue='origin')   #dodge separates hues in x

plt.legend(loc=1)
plt.show()
```

##**Heatmap**

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm',annot=True)
plt.show()
```

```python
#Heatmap with Pivot Table

year=df.pivot_table(index='origin', columns='model_year' , values='mpg')
sns.heatmap(year , cmap='coolwarm' , annot=True ,linecolor='white', linewidth='1')
```

**##Facet grids**

```python
df.head()
```

```python
df['origin'].value_counts()
```

```python
g = sns.FacetGrid(df, col="cylinders", row='origin')
g = g.map(plt.scatter, "mpg", "weight") # Apply a plotting function to each facet's subset of the data.
```

```python
g = sns.FacetGrid(df, col="cylinders", hue='origin')
g = g.map(plt.scatter, "mpg", "weight").add_legend()
```
