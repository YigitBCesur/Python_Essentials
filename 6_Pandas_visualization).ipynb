# Pandas for visualization


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

## The Data

There are some fake data csv files you can read in as dataframes:

```python
df = yf.download('AAPL',start='2022-01-01', end="2023-10-28")
df.head()
```

```python
df['return']= df['Adj Close'].pct_change(1)
df.head()
```

```python
df.drop(df.columns[[1,2,3,5]], axis=1, inplace=True)
```

```python
df=df.drop(df.index[0])
df
```

```python
df.info()
```

```python
df['Adj Close'].hist()
plt.show()
```

```python
df['return'].hist(bins=30)
plt.show()
```

```python
df['return'].plot(kind="hist", bins=30)
plt.show()
```

There are many different plot types in pandas, try to explore the followings on your own:

* df.plot.area  
* df.plot.pie
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box        
* df.plot.kde  
* df.plot.density


You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above
___

```python
df.plot(subplots=True)
plt.show()
```

```python
df['return'].plot()   # by default the kind ='line'
plt.axhline(0, color ='red', lw = 2)
plt.show()
```

```python
df.plot.scatter(x='return',y='Adj Close')
plt.show()
```

```python
df[['Adj Close','return']].plot.box()
plt.show()
```

```python
# Kernel Density Estimation plot (KDE)
df['return'].plot.kde()
plt.show()
```

```python
plt.figure(figsize=(14,8))
df[['Open', 'Adj Close']].plot(kind='area' , stacked=True , alpha=0.50 , title='Chart' , legend=True);   #alpha=shading
#df[['Open', 'Adj Close']].abs().plot(kind='area' , stacked=True)    #with absolute numbers


#If we dont have any negatis value or No need to normalization in our dataset
#df.plot(kind='area' , stacked=False)
```
