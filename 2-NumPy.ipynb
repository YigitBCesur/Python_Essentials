## 5. Boolean Indexing

```python
a = np.arange(1,11) #array from 1 to 10
a
```

```python
mask1 = a > 5 #element-wise check if greater than 5
mask1
```

```python
mask2 = a < 8 #element-wise check if smaller than 8
mask2
```

```python
mask3 = (a > 5) & (a < 8) #element-wise check if greater 5 and smaller 8 (logical and)
mask3
```

```python
mask4 = (a > 5) | (a < 8) #element-wise check if greater 5 or smaller 8 (logical or)
mask4
```

```python
mask5 = ~((a > 5) & (a < 8)) #the opposite of mask3
mask5
```

```python
a[a>5] #slicing all elements that are greater 5 (fulfill condition of mask1)
```

```python
a[mask1] #slicing all elements that fulfill condition of mask1
```

```python
a[mask2] #slicing all elements that fulfill condition of mask2
```

```python
a[mask3] #slicing all elements that fulfill condition of mask3
```

```python
a[mask4] #slicing all elements that fulfill condition of mask4
```

```python
a[mask5] #slicing all elements that fulfill condition of mask5
```

## 8. Axis Logic
When working with 2-dimensional arrays (matrices) we have to consider rows and columns. This becomes very important when we get to the section on pandas. In array terms, axis 0 is the vertical axis (elements on the rows), and axis 1 is the horizonal axis (elements on the columns). These values (0,1) correspond to the order in which <tt>arr.shape</tt> values are returned.

Let's see how this affects our summary statistic calculations from above.

```python
my_matrix
```

```python
# axis 0 (zero) is doing a vertical operation, and axis 1 a horizonal one.
# again, note that the logic is different from R, Matlab or etc
my_matrix.sum(axis=0)
```

```python
my_matrix.sum(1)
```

```python
my_matrix * my_matrix  # note that this is not a matrix multiplication
```

```python
# To do a matrix multiplication use np.dot (equivalent to %*% in R)
np.dot(my_matrix,my_matrix.T)
```

```python
# inverse of a matrix
A = np.array([[1,2],[3,4]])
A
```

```python
A_inv = np.linalg.inv(A)
A_inv
```
