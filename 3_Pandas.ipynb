## 1. Series

Series are very similar to NumPy arrays. The difference is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. We can convert a list,numpy array, or dictionary to a Series.

NumPy arrays are like lists, but they can only store one type of data, such as all numbers or all strings. They can also be very fast at performing mathematical operations on their data.

Pandas Series are like NumPy arrays, but they can store any type of data, and they have labels associated with each value. This makes them easier to read and understand, and it also allows you to perform more complex data analysis operations, such as filtering and sorting.

```python
import numpy as np
import pandas as pd
```

```python
data = [10,20,30,40]
data
```

```python
pd.Series(data=data)
```

```python
labels = ['alice','bob','carroll','dave']
pd.Series(data=data,index=labels)
```

```python
my_arr = np.array([4,5,6,7])
pd.Series(my_arr)
```

```python
pd.Series(my_arr,labels)
```

```python
my_dict = {'row1':15,'row2':25,'row3':35}
pd.Series(my_dict)
```

```python
my_series = pd.Series(my_dict)
my_series
```

```python
my_series[0]         # unlike dictionaries, we can extract info by index number and label.
```

```python
my_series['row3']
```

#### Conditional mutation (creating new variables)
Let's say we want to create a new variable named "is_pass" and is defined as "is_pass"=1 if "speed">70 and 0 otherwise. There are multiple ways to do that:

```python
# let's begin with the hard way!
df2['speed']>70
```

```python
temp=[]
for x in df2['speed']:
    if x>70: temp.append('fast')
    else: temp.append('slow')

df2['is_fast']=temp
df2
```

Can we do it in one line? How about using list comprehension?

```python
df2['is_faster'] = ['unacceptable' if x > 70 else 'acceptable' for x in df2['speed']]

df2
```

Is there any other way?

```python
df2.loc[df2['speed']> 80, 'is_pass3'] ='jail'
df2.loc[df2['speed'].between(70, 80), 'is_pass3'] = 'ticket'
df2.loc[df2['speed']<=70, 'is_pass3'] ='okay'
df2.head()
```

Is there any other way?

```python
df2['is_pass4']=np.where(df2['speed']>70, 'ticket','warning')
df2.head()
```
